# PHP - Pulsating Heat Pipe [AL203, 60FR, Heat Inputs]

```python
from PyPulseHeatPipe import PulseHeatPipe
from PyPulseHeatPipe import DataVisualisation
```

```python
d_40w = 'data/nf_heat_inputs_6ofr/40w/'
s_40w = 'nf_60fr_40w'
d_60w = 'data/nf_heat_inputs_6ofr/60w/'
s_60w = 'nf_60fr_60w'
d_80w = 'data/nf_heat_inputs_6ofr/80w/'
s_80w = 'nf_60fr_80w'
d_100w = 'data/nf_heat_inputs_6ofr/100w/'
s_100w = 'nf_60fr_100w'
d_120w = 'data/nf_heat_inputs_6ofr/120w/'
s_120w = 'nf_60fr_120w'
```

```python
analysis_40w = PulseHeatPipe(datapath=d_40w, sample=s_40w)
analysis_60w = PulseHeatPipe(datapath=d_60w, sample=s_60w)
analysis_80w = PulseHeatPipe(datapath=d_80w, sample=s_80w)
analysis_100w = PulseHeatPipe(datapath=d_100w, sample=s_100w)
analysis_120w = PulseHeatPipe(datapath=d_120w, sample=s_120w)
```

```python
df_40w, df_conv_40w = analysis_40w.data_etl(name='nf_*')
df_60w, df_conv_60w = analysis_60w.data_etl(name='nf_*')
df_80w, df_conv_80w = analysis_80w.data_etl(name='nf_*')
df_100w, df_conv_100w = analysis_100w.data_etl(name='nf_*')
df_120w, df_conv_120w = analysis_120w.data_etl(name='nf_*')
```
